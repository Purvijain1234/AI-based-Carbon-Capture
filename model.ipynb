```python

import pandas as pd
import xgboost as xgb
from datetime import datetime, timedelta
import json

# Load trained model
model = xgb.XGBRegressor()
model.load_model("ai_model.json")

# Generate dates
start_date = datetime(2025, 1, 1)
num_days = 300   # generate for 30 days

dates = [start_date + timedelta(days=i) for i in range(num_days)]

# Create feature dataframe
df = pd.DataFrame({
    "dayofyear": [d.timetuple().tm_yday for d in dates],
    "dayofweek": [d.weekday() for d in dates],
    "quarter": [(d.month - 1) // 3 + 1 for d in dates],
    "month": [d.month for d in dates],
    "year": [d.year for d in dates]
})

# Predict CO2
predictions = model.predict(df)

# Build JSON
output = {
    "model": "XGBoost COâ‚‚ Predictor",
    "unit": "ppm",
    "decision_threshold": 420,
    "predictions": {}
}

for d, p in zip(dates, predictions):
    output["predictions"][d.strftime("%Y-%m-%d")] = round(float(p), 2)

# Save
with open("ai_predictions.json", "w") as f:
    json.dump(output, f, indent=2)

print("AI predictions exported successfully")
```
